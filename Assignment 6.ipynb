```python
import pandas as pd
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

Now we will download the webDriver for the Web Browser. Steps for download are; 1) Check the version of your browser 2) go to the link https://chromedriver.chromium.org/downloads 3) Download the webdriver for your version of your browser

```python
driver=webdriver.Chrome()
```

```python
#Opening the shine page on automated chrome browser
driver.get('https://www.shine.com')
```

```python
#entering designation and location as required in the question
designation=driver.find_element(By.CLASS_NAME,'form-control')
designation.send_keys('Data Analyst')
```

```python
location=driver.find_element(By.XPATH,'/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[1]/ul/li[2]/div/input')
location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.XPATH,"/html/body/div/div[4]/div/div[2]/div[2]/div/form/div/div[2]/div/button")
search.click()
```

```python
job_title=[]
job_location=[]
experience_required=[]
company_name=[]
```

```python
title_tags=driver.find_elements(By.XPATH,'//h2[@itemprop="name"]/a')
for i in title_tags[0:10]:
    title=i.text
    job_title.append(title)
    
location_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
for i in location_tags[0:10]:
    location=i.text
    job_location.append(location)
    
experience_tags=driver.find_elements(By.XPATH,'//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_jobIcon__3FB1t"]')
for i in experience_tags[0:10]:
    experience=i.text
    experience_required.append(experience)
    
company_tags=driver.find_elements(By.XPATH,'//div[@class="jobCard_jobCard_cName__mYnow"]')
for i in company_tags[0:10]:
    company=i.text
    company_name.append(company)
```

```python
print(len(job_title),len(job_location),len(experience_required),len(company_name))
```

```python
import pandas as pd
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Experience':experience_required,'Company':company_name})
df
```

```python
import pandas as pd
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.flipkart.com')
```

```python
brands=driver.find_element(By.CLASS_NAME,'Pke_EE')
brands.send_keys('sunglasses')
```

```python
search=driver.find_element(By.CLASS_NAME,"_2iLD__")
search.click()
```

```python
Product_Brand=[]
Product_Description=[]
Product_Price=[]
Product_Offer=[]
```

```python
start=0
end=3
for pages in range(start,end):
    Brand_tags=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in Brand_tags[0:100]:
    Brand=i.text
    Product_Brand.append(Brand)
    
next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
next_button.click()
time.sleep(3)

start=0
end=3
for pages in range(start,end):
    Description_tags=driver.find_elements(By.XPATH,'//div[@class="_2B099V"]/a[1]')
for i in Description_tags[0:100]:
    Description=i.text
    Product_Description.append(Description)
    
next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
next_button.click()
time.sleep(3)

start=0
end=3
for pages in range(start,end):
    Price_tags=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
for i in Price_tags[0:100]:
    Price=i.text
    Product_Price.append(Price)
    
next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
next_button.click()
time.sleep(3)

start=0
end=3
for pages in range(start,end):
    Offer_tags=driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
for i in Offer_tags[0:100]:
    Offer=i.text
    Product_Offer.append(Offer)
    
next_button=driver.find_element(By.XPATH,'/html/body/div/div/div[3]/div[1]/div[2]/div[12]/div/div/nav/a[11]')
next_button.click()
time.sleep(3)
```

```python
print(len(Product_Brand),len(Product_Description),len(Product_Price),len(Product_Offer))
```

```python
import pandas as pd
df=pd.DataFrame({'BRAND':Product_Brand,'DESCRIPTION':Product_Description,'PRICE':Product_Price,'OFFER':Product_Offer})
df
```
