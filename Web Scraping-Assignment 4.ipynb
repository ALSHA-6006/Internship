```python
import pandas
import selenium
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
Name_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
for i in Name_tags:
    Name.append(i.text)

Artist_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
for i in Artist_tags:
    Artist.append(i.text)

Upload_date_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
for i in Upload_date_tags:
    Upload_date.append(i.text)
    
Views_tags=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
for i in Views_tags:
    Views.append(i.text)
```

```python
print(len(Name),len(Artist),len(Upload_date),len(Views))
```

```python
import pandas as pd
df=pd.DataFrame({'NAME':Name,'ARTIST':Artist,'UPLOAD_DATE':Upload_date,'VIEWS':Views})
df
```

```python
Dataset_name=[]
Data_type=[]
Task=[]
Attribute_Type=[]
No_of_instances=[]
No_of_attribute=[]
Year=[]
```

```python
print(len(Dataset_name),len(Data_type),len(Task),len(Attribute_Type),len(No_of_instances),len(No_of_attribute),len(Year))
```

```python
import pandas as pd
df=pd.DataFrame({'DATASET_NAME':Dataset_name,'DATA_TYPE':Data_type,'TASK':Task,'ATTRIBUTE_TYPE':Attribute_Type,'NO_OF_INSTANCES':No_of_instances,'NO_OF_ATTRIBUTE':No_of_attribute,'YEAR':Year})
df
```
